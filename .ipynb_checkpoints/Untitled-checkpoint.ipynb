```julia
B=Fi1 
A0=SIGMA 
ND=1
N=4 
L=2
HORZ=8

B = convert(Array{Float64}, varcompanion(B, ND, N, L))
	J = convert(Array{Float64}, [Diagonal(ones(N)) zeros(N, N*(L-1))])

	ir1 = Array{Float64}(undef, N, N, HORZ+1);
	wold = Array{Float64}(undef, N, N, HORZ+1);

	# GET MA coefficients
	jT = J'
	bh = B ^ 0

	 for h in 0:HORZ
		wold[:, :, h+1] = J * ((bh) * jT)
		bh = bh * B
	end

	A0 = (1.0 ./ sqrt.(diag(A0))) .* A0

   for h in 1:(HORZ+1)
		for i in 1:N
			ir1[:, i, h] = A0[i,:]'*wold[:,:,h]';
		end
	end
```
